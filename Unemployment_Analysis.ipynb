```python
from google.colab import files
import pandas as pd

uploaded = files.upload()  # Prompts you to upload a file

# Assuming you uploaded a file named 'data.csv'
df = pd.read_csv('Unemployment_in_India.csv')
print(df.head())
```

```python
# Convert 'Date' column to datetime
df[' Date'] = pd.to_datetime(df[' Date'].str.strip(), format='%d-%m-%Y')
df[' Date']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Line plot of unemployment rate over time for a specific region (e.g., Gujarat)
Gujarat_data = df[df['Region'] == 'Gujarat']

plt.figure(figsize=(12, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=Gujarat_data, hue='Area')
plt.title('Unemployment Rate Over Time in Gujrat')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Strip column names of extra spaces
df.columns = df.columns.str.strip()

plt.figure(figsize=(12, 8))
sns.lineplot(
    x='Region',
    y='Estimated Unemployment Rate (%)',
    data=df,
    hue='Area',
    ci=None  # <- This removes the shadow
)
plt.title('Lineplot of Unemployment Rate by Region and Area')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

if df.empty:
    print("The dataframe is empty!")
else:
    df_numeric = df.select_dtypes(include=['number'])
    if df_numeric.isnull().any().any():
        print("Missing values detected. Handling missing values...")
        df_numeric = df_numeric.dropna()

    correlation_matrix = df_numeric.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()
```

```python
sns.countplot(data=df, x='Area')
plt.title('Distribution of Area')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = df.dropna(subset=['Estimated Unemployment Rate (%)', 'Area'])
df['Area'] = df['Area'].astype('category')
sns.histplot(data=df, x='Estimated Unemployment Rate (%)', kde=True, hue='Area')
plt.title('Distribution of Estimated Unemployment Rate (%)')
plt.show()
```
